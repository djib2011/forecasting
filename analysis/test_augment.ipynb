```python
import pickle as pkl
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('../data/yearly_18_train.pkl', 'rb') as f:
    data = pkl.load(f)
```

```python
data[0].shape
```

```python
data[1].shape
```

```python
num_synthetic_samples = 500000
```

```python
s1 = np.random.choice(np.arange(len(data[0])), num_synthetic_samples, replace=True)
s2 = np.random.choice(np.arange(len(data[0])), num_synthetic_samples, replace=True)
```

```python
syn = [(d[s1] + d[s2]) / 2 for d in data]
```

```python
plt.plot(data[0][s1[0]])
plt.plot(data[0][s2[0]])
plt.plot(syn[0][0])
```

```python
plt.plot(data[1][s1[0]])
plt.plot(data[1][s2[0]])
plt.plot(syn[1][0])
```

```python
plt.plot(data[0][s1[1]])
plt.plot(data[0][s2[1]])
plt.plot(syn[0][1])
```

```python
plt.plot(data[1][s1[1]])
plt.plot(data[1][s2[1]])
plt.plot(syn[1][1])
```
